```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
```

```python
ds=pd.read_csv("C:\\Users\\ELCOT\\Desktop\\archive\\gpa_study_hours.csv")
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
#to check linearity

plt.scatter(ds['study_hours'], ds['gpa'], color='brown')
plt.xlabel('Hours of Study')
plt.ylabel('GPA')
plt.title('Linearity of Study Hours with GPA')
plt.show()
```

```python
X = ds[['study_hours']]
y = ds['gpa']
```

```python
# Splitting hte data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
# Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Evaluate model performance
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
mse,mae
```

```python
# Plotting the linear regression line and scatter plot
plt.scatter(X_test, y_test, color='orange', label='Actual')
plt.plot(X_test, y_pred, color='blue', label='Predicted')
plt.xlabel('Study Hours')
plt.ylabel('GPA')
plt.title('Relation between Study Hours and GPA')
plt.legend()
plt.show()
```

```python
print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R-squared: {r2:.2f}")
```

```python
normalized_mae = mean_absolute_error(y_test, y_pred) / (max(y) - min(y))
normalized_rmse = np.sqrt(mean_squared_error(y_test, y_pred)) / (max(y) - min(y))

print(f"Normalized MAE: {normalized_mae:.2f}")
print(f"Normalized RMSE: {normalized_rmse:.2f}")
```
